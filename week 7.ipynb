```python
import pandas as pd 


data = pd.read_csv('student_habits_performance.csv')

print(data)
```

```python
data.head()
```

```python
data.describe()
```

```python
data['diet_quality'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['exercise_frequency'], data['diet_quality'])
plt.xlabel('exercise_frequency')
plt.ylabel('diet_quality')
plt.title('Diet vs Excercise frequency')
plt.show()
```

```python
data.groupby('parental_education_level')['age'].mean().plot(kind='bar')
plt.xlabel('parental_education_level')
plt.ylabel('Average age')
plt.title('Average age by Education level')
plt.show()
```

```python
data['exam_score'].plot(kind='hist', bins=10)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
data.plot(kind='scatter', x='attendance_percentage', y='exam_score')
plt.title('Attendance vs Exam Score')
plt.show()
```
